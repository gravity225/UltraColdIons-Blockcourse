# Calibrationcurve plot

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Times New Roman']
plt.rcParams['font.size'] = 16


# Replace with your actual Parquet file path
file_path = 'C:\\Users\\samse\\Desktop\\Blockkurs\\07-UltracoldIons\\Repository\\Datenanalysis_PSD\\ParquetData\\calibrationcurve_fixed_frequency.parquet'

# Read the Parquet file
df = pd.read_parquet(file_path)

# Display the first few rows to confirm
print(df.head())

# Assume the first two columns are x and y
x_column = df.columns[0]
y_column = df.columns[1]

# Plot the data
plt.figure(figsize=(20, 5))
plt.plot(df[x_column], df[y_column], color='#006400')
plt.xlabel("Time [s]")
plt.ylabel("Amplitude [V]")
plt.title('Temperature Fluctuation of OFD')
plt.grid(True)
plt.xlim(-100,100)


plt.savefig("C:\\Users\\samse\\Desktop\\Blockkurs\\07-UltracoldIons\\Bericht\\ALLPLOTS_SVG\\Fluctuation.svg", dpi=300, bbox_inches='tight')
plt.savefig("C:\\Users\\samse\\Desktop\\Blockkurs\\07-UltracoldIons\\Bericht\\ALLPLOTS_PNG\\Fluctuation.png", dpi=300, bbox_inches='tight')
plt.show()
```
